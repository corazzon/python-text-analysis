### N-grams 

* í† í°ì„ ëª‡ ê°œ ì‚¬ìš©í•  ê²ƒì¸ì§€ë¥¼ êµ¬ë¶„í•©ë‹ˆë‹¤. ì§€ì •í•œ nê°œì˜ ìˆ«ì ë§Œí¼ì˜ í† í°ì„ ë¬¶ì–´ì„œ ì‚¬ìš©í•©ë‹ˆë‹¤.
* ì˜ˆë¥¼ ë“¤ì–´ (1, 1) ì´ë¼ë©´ 1ê°œì˜ í† í°ì„ (2, 3)ì´ë¼ë©´ 2~3ê°œì˜ í† í°ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.
* analyzer ì„¤ì •ì— ë”°ë¼ ë‹¨ì–´ë‹¨ìœ„, ìºë¦­í„° ë‹¨ìœ„ì— ë”°ë¼ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

* ê¸°ë³¸ê°’ = (1, 1)
* ngram_range(min_n, max_n)
* min_n <= n <= max_n
```
(1, 1) ì€ 1 <= n <= 1
(1, 2) ì€ 1 <= n <= 2
(2, 2) ì€ 2 <= n <= 2
```

````python
# ë‹¨ì–´ê°€ ë„ˆë¬´ ë§ì•„ì„œ ì¶œë ¥ì´ ì˜¤ë˜ ê±¸ë¦°ë‹¤ë©´ max_columns ê°’ì„ ì¡°ì •í•´ì„œ ì‚¬ìš©í•©ë‹ˆë‹¤.
pd.options.display.max_columns = None
````

````python
# ngram_range: ì¶”ì¶œí•  ë‹¤ë¥¸ ë‹¨ì–´ n-gram ë˜ëŠ” char n-gramì— ëŒ€í•œ n-ê°’ ë²”ìœ„ì˜ í•˜í•œ ë° ìƒí•œì…ë‹ˆë‹¤. ê¸°ë³¸ê°’ = (1, 1)
# ngram_range=(1, 2)
# help(CountVectorizer)
cvect = CountVectorizer(ngram_range=(1, 2))
X = cvect.fit_transform(corpus)
dtm = X.toarray()
dtm
````

````python
# df_dtm.sum ìœ¼ë¡œ ë¹ˆë„ìˆ˜ í•©ê³„ êµ¬í•˜ê¸°
pd.DataFrame(dtm, columns=cvect.get_feature_names_out()).style.background_gradient()
````

````python
# ëª¨ë¸ì„ ë°›ì•„ ë³€í™˜ì„ í•˜ê³  ë¬¸ì„œ ìš©ì–´ í–‰ë ¬ì„ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“¤ì–´ ì¬ì‚¬ìš©í•©ë‹ˆë‹¤.
def display_transform_dtm(cvect, corpus):
    """
    ëª¨ë¸ì„ ë°›ì•„ ë³€í™˜ì„ í•˜ê³  ë¬¸ì„œ ìš©ì–´ í–‰ë ¬ì„ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜
    """
    X = cvect.fit_transform(corpus)
    print(cvect.get_feature_names_out())
    dtm = X.toarray()
    df_dtm = pd.DataFrame(dtm, columns=cvect.get_feature_names_out()).style.background_gradient()
    return df_dtm
````

## TF-IDF(Term Frequency-Inverse Document Frequency)
TF(ë‹¨ì–´ ë¹ˆë„, term frequency)ëŠ” íŠ¹ì •í•œ ë‹¨ì–´ê°€ ë¬¸ì„œ ë‚´ì— ì–¼ë§ˆë‚˜ ìì£¼ ë“±ì¥í•˜ëŠ”ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ê°’ìœ¼ë¡œ, ì´ ê°’ì´ ë†’ì„ìˆ˜ë¡ ë¬¸ì„œì—ì„œ ì¤‘ìš”í•˜ë‹¤ê³  ìƒê°í•  ìˆ˜ ìˆë‹¤. í•˜ì§€ë§Œ ë‹¨ì–´ ìì²´ê°€ ë¬¸ì„œêµ° ë‚´ì—ì„œ ìì£¼ ì‚¬ìš©ë˜ëŠ” ê²½ìš°, ì´ê²ƒì€ ê·¸ ë‹¨ì–´ê°€ í”í•˜ê²Œ ë“±ì¥í•œë‹¤ëŠ” ê²ƒì„ ì˜ë¯¸í•œë‹¤. ì´ê²ƒì„ DF(ë¬¸ì„œ ë¹ˆë„, document frequency)ë¼ê³  í•˜ë©°, ì´ ê°’ì˜ ì—­ìˆ˜ë¥¼ IDF(ì—­ë¬¸ì„œ ë¹ˆë„, inverse document frequency)ë¼ê³  í•œë‹¤. TF-IDFëŠ” TFì™€ IDFë¥¼ ê³±í•œ ê°’ì´ë‹¤.

IDF ê°’ì€ ë¬¸ì„œêµ°ì˜ ì„±ê²©ì— ë”°ë¼ ê²°ì •ëœë‹¤. ì˜ˆë¥¼ ë“¤ì–´ 'ì›ì'ë¼ëŠ” ë‚±ë§ì€ ì¼ë°˜ì ì¸ ë¬¸ì„œë“¤ ì‚¬ì´ì—ì„œëŠ” ì˜ ë‚˜ì˜¤ì§€ ì•Šê¸° ë•Œë¬¸ì— IDF ê°’ì´ ë†’ì•„ì§€ê³  ë¬¸ì„œì˜ í•µì‹¬ì–´ê°€ ë  ìˆ˜ ìˆì§€ë§Œ, ì›ìì— ëŒ€í•œ ë¬¸ì„œë¥¼ ëª¨ì•„ë†“ì€ ë¬¸ì„œêµ°ì˜ ê²½ìš° ì´ ë‚±ë§ì€ ìƒíˆ¬ì–´ê°€ ë˜ì–´ ê° ë¬¸ì„œë“¤ì„ ì„¸ë¶„í™”í•˜ì—¬ êµ¬ë¶„í•  ìˆ˜ ìˆëŠ” ë‹¤ë¥¸ ë‚±ë§ë“¤ì´ ë†’ì€ ê°€ì¤‘ì¹˜ë¥¼ ì–»ê²Œ ëœë‹¤.

ì—­ë¬¸ì„œ ë¹ˆë„(IDF)ëŠ” í•œ ë‹¨ì–´ê°€ ë¬¸ì„œ ì§‘í•© ì „ì²´ì—ì„œ ì–¼ë§ˆë‚˜ ê³µí†µì ìœ¼ë¡œ ë‚˜íƒ€ë‚˜ëŠ”ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ê°’ì´ë‹¤. ì „ì²´ ë¬¸ì„œì˜ ìˆ˜ë¥¼ í•´ë‹¹ ë‹¨ì–´ë¥¼ í¬í•¨í•œ ë¬¸ì„œì˜ ìˆ˜ë¡œ ë‚˜ëˆˆ ë’¤ ë¡œê·¸ë¥¼ ì·¨í•˜ì—¬ ì–»ì„ ìˆ˜ ìˆë‹¤.

* ì¶œì²˜ : [TF-IDF - ìœ„í‚¤ë°±ê³¼, ìš°ë¦¬ ëª¨ë‘ì˜ ë°±ê³¼ì‚¬ì „](https://ko.wikipedia.org/wiki/TF-IDF)

\begin{equation*}

### IDF(Inverse Document Frequency)

- IDF ê°’ì€ ë¬¸ì„œêµ°ì˜ ì„±ê²©ì— ë”°ë¼ ê²°ì •ëœë‹¤. ì˜ˆë¥¼ ë“¤ì–´ 'ì›ì'ë¼ëŠ” ë‚±ë§ì€ ì¼ë°˜ì ì¸ ë¬¸ì„œë“¤ ì‚¬ì´ì—ì„œëŠ” ì˜ ë‚˜ì˜¤ì§€ ì•Šê¸° ë•Œë¬¸ì— IDF ê°’ì´ ë†’ì•„ì§€ê³  ë¬¸ì„œì˜ í•µì‹¬ì–´ê°€ ë  ìˆ˜ ìˆì§€ë§Œ, ì›ìì— ëŒ€í•œ ë¬¸ì„œë¥¼ ëª¨ì•„ë†“ì€ ë¬¸ì„œêµ°ì˜ ê²½ìš° ì´ ë‚±ë§ì€ ìƒíˆ¬ì–´ê°€ ë˜ì–´ ê° ë¬¸ì„œë“¤ì„ ì„¸ë¶„í™”í•˜ì—¬ êµ¬ë¶„í•  ìˆ˜ ìˆëŠ” ë‹¤ë¥¸ ë‚±ë§ë“¤ì´ ë†’ì€ ê°€ì¤‘ì¹˜ë¥¼ ì–»ê²Œ ëœë‹¤.


$IDF(t) = \log \frac{N}{df(t)}$

* ê³¼ë„í•œ ë¹ˆë„ ê°ì†Œ: ë§¤ìš° í”í•˜ê²Œ ë‚˜íƒ€ë‚˜ëŠ” ë‹¨ì–´ë“¤ì€ ë‚®ì€ IDF ê°’ì„ ê°–ê²Œ ë©ë‹ˆë‹¤. ì´ëŠ” ê³¼ë„í•œ ë¹ˆë„ë¥¼ ê°€ì§„ ì¼ë°˜ì ì¸ ë‹¨ì–´ë“¤ì´ ê²€ìƒ‰ì´ë‚˜ ë¶„ë¥˜ì—ì„œ í° ì—­í• ì„ í•˜ì§€ ëª»í•˜ë„ë¡ ë°©ì§€í•©ë‹ˆë‹¤.

````python
# í•˜ë‚˜ì˜ ë¬¸ì„œì—ë§Œ ë‚˜íƒ€ë‚˜ëŠ” í† í°ì€ idf ê°€ì¤‘ì¹˜ê°€ ë†’ìŠµë‹ˆë‹¤.
# ì ê²Œ ë‚˜íƒ€ë‚œ í† í°ì´ë¼ë„ ëª¨ë“  ë¬¸ì„œì—ë„ ìˆëŠ” í† í°ì€ idfê°€ ë‚®ìŠµë‹ˆë‹¤.
idf = tfidfvect.idf_
idf
````
